```python
fp=open('/content/drive/MyDrive/지역평균기온.txt','r',encoding='utf-8')
data=fp.readlines()
print(type(data))
data
fp.close()
```

```python
dt={}#  빈 딕셔너리만들기
for line in data:
  line = line.replace("\n","")
  #print(line)
  items=line.split(',')
  #print(items)
  dt[items[0]]=int(items[1])
print(type(dt))
dt
```

```python
x=list(dt.keys())
y=list(dt.values())
#print(type(x))
#print(type(y))
plt.figure(figsize=(20,12))
plt.subplot(3,2,1)
plt.xlabel('도시')
plt.ylabel('온도')
for i in range(0,len(x)):
  plt.text(i,y[i],y[i],ha='center')
plt.plot(x,y,'go--')
plt.subplot(3,2,2)
plt.xlabel('도시')
plt.ylabel('온도')
for i in range(0,len(x)):
  plt.text(i,y[i],y[i],ha='center')
plt.scatter(x,y)
plt.subplot(3,2,3)
plt.xlabel('도시')
plt.ylabel('온도')
for i in range(0,len(x)):
  plt.text(i,y[i],y[i],ha='center')
plt.bar(x,y)
plt.subplot(3,2,4)
plt.xlabel('도시')
plt.ylabel('온도')
plt.barh(x,y)
```
